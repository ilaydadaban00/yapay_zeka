```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Rastgele bir veri kmesi oluturuyoruz
np.random.seed(0)
X1 = np.random.normal(loc=4, scale=1, size=(100, 2))  # Kme 1
X2 = np.random.normal(loc=10, scale=1, size=(100, 2)) # Kme 2
X3 = np.random.normal(loc=15, scale=1, size=(100, 2)) # Kme 3

# Verileri birletiriyoruz
X = np.concatenate([X1, X2, X3], axis=0)

# K-means ile kmeleme
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)

# Sonular grselletirme
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', marker='X')
plt.title("K-means Kmeleme Sonular")
plt.xlabel('zellik 1')
plt.ylabel('zellik 2')
plt.show()
```
