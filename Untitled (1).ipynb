```python
import pandas as pd
data={
    'year':[2020,2021,2022,2023,2024],
    
    'sales (USD)':[50000,75000,100000,12000,15000],
    'customers':[120,180,250,300,400]
}
df=pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(12,6))
```

```python
plt.subplot(1,2,1)
plt.plot(df['year'],df['sales (USD)'],marker='o',color="blue")
plt.title('yÄ±llÄ±k satÄ±slar')
plt.xlabel("yÄ±l")
plt.ylabel('satÄ±ÅŸ (USD)')
plt.grid()
plt.subplot(1,2,2)
plt.bar(df['year'],df['customers'], color='orange')
plt.title('yÄ±llÄ±k musteri sayÄ±sÄ±')
plt.xlabel('yÄ±l')
plt.ylabel('musteri sayÄ±sÄ±')
```
